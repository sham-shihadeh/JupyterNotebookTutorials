## Abstract
**Exploratory Data Analysis (EDA)** is a method for analyzing and summarizing essential dataset characteristics. It provides insights into data distribution and the presence of null values, utilizing both graphical (Matplotlib) and non-graphical techniques. In the context of Machine Learning, EDA helps assess data quality and the potential performance of ML models.

**Data visualization** is a fundamental component of modern data analysis and interpretation. It involves representing data through graphics such as charts, plots, infographics, and sometimes animations. This type of visualization simplifies the understanding of data relationships, making it effective for decision-making, research, and analysis.

The ability to transform raw data into meaningful insights is why data visualization is important. It enables individuals to discern patterns, trends, and anomalies in datasets, allowing for better-informed decisions and a deeper understanding of complex datasets.

## Exploratory Data Analysis (EDA) using Pandas
Pandas is a versatile Python library essential for effective data handling and analysis in EDA. For example it is used in: 

1. **Data Loading**: Create DataFrames efficiently from dictionaries, showcasing Pandas' data structuring capabilities.

2. **Data Display**: Easily view, select, and filter data, like choosing the 'Age' column or filtering for specific criteria.

3. **Data Manipulation:** Add columns, perform calculations (e.g., average age), and sort data as needed.

4. **Grouping and Aggregation**: Group data by attributes (e.g., 'City') and compute aggregations (e.g., average age per city).

5. **Concatenation**: Seamlessly merge and extend datasets using the concat function.

>Pandas Empowerment: Pandas' data structures, including DataFrames and Series, facilitate data exploration, cleaning, transformation, and aggregation, enabling analysts and data scientists to uncover patterns, anomalies, and valuable insights in EDA.

***

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {'Name': ['Kevin', 'Michael', 'Jim', 'Dwight'],
        'Age': [40, 54, 76, 87],
        'City': ['London', 'Leicester', 'Scranton', 'New York']}
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:\n", df, "\n")
print("-----------------------")

# Selecting columns
ages = df['Age']
print("Ages:\n", ages, "\n")
print("-----------------------")

# Filtering rows
young_people = df[df['Age'] < 30]
print("Young People:\n", young_people, "\n")
print("-----------------------")

# Adding a new column
df['Salary'] = [60000, 75000, 50000, 80000]
print("Salary in the DataFrame:\n", df, "\n")
print("-----------------------")

# Calculating statistics
average_age = df['Age'].mean()
print("Average Age:", average_age, "\n")
print("-----------------------")

# Sorting by a column
sorted_df = df.sort_values(by='Age')
print("Sorted by Age:\n", sorted_df, "\n")
print("-----------------------")

# Grouping and aggregation
city_groups = df.groupby('City')
city_stats = city_groups['Age'].mean()
print("Average Age by City:\n", city_stats, "\n")
print("-----------------------")

# Creating a new DataFrame
new_data = {'Name': ['Pam', 'Kelly'],
            'Age': [26, 45],
            'City': ['Istanbul', 'Washington']}
new_df = pd.DataFrame(new_data)

# Concatenating DataFrames
combined_df = pd.concat([df, new_df], ignore_index=True)
print("Combined DataFrame:\n", combined_df)
```

## Matplotlib for visualising data
Matplotlib is a powerful Python library designed for creating static Matplotlib is a powerful Python library designed for creating static data visualizations. It provides a flexible framework for generating an array of graphs, charts, and plots that closely resemble the graphics produced by MATLAB. Matplotlib is built on top of NumPy, a fundamental Python library for numerical operations, making it an essential tool for data analysis and visualization.

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Import the Matplotlib library for plotting
import numpy as np  # Import the NumPy library for numerical operations
```

***
#### Line Plot 
> Matplotlib can be used to create line plots to visualize trends in time-series data, such as stock prices over time or sales data. For example:

```python
# Generate sample data
x = np.linspace(0, 10, 100)  # Create an array 'x' with 100 equally spaced values from 0 to 10
y = np.sin(x)  # Calculate the sine values for each 'x' value and store them in 'y'

# Create a line plot
plt.plot(x, y)  # Create a line plot with 'x' on the x-axis and 'y' on the y-axis
plt.xlabel('Time')  # Set the x-axis label as 'Time'
plt.ylabel('Value')  # Set the y-axis label as 'Value'
plt.title('Line Plot Example')  # Set the title of the plot as 'Example Line Plot'
plt.show()  # Display the plot on the screen
```

***
#### Bar Chart:
> Bar charts are used for comparing categorical data, such as the performance of sales of different products. For example:

```python
# Sample data
categories = ['A', 'B', 'C']
values = [20, 35, 28]

# Create a bar chart
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example ')
plt.show()
```

***
#### Scatter Plot
> Scatter plots are used to show the relationship between two continuous variables and can be used to examine correlations or clusters in data.

```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot Example')
plt.show()
```

***
#### Histagrams 
> Histograms are beneficial for showing the distribution of a single variable. Matplotlib is used to analyze the distribution of data, such as exam scores. For example:

```python
# Generate random data
data = np.random.normal(0, 1, 1000)

# Create a histogram
plt.hist(data, bins=20, edgecolor='k')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

***

***
#### Pie Chart
> Pie Charts are useful for showing parts of a whole, for example: 

```python
# Sample data
labels = ['A', 'B', 'C']
sizes = [20, 30, 20]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

>The above examples showcase some of the common use cases for Matplotlib in visualizing data. They can all be customised to suit the specific needs of each use case.

## Advanced Visualisation using Seaborn
> the code in this section is based on Source 5

Seaborn is another Python library based on Matplotlib that is used for creating statistical graphics. The example below demonstrates the **relplot** method, which is more useful than using Matplotlib because you provide the variable names and their intended roles in the plot. 

Seaborn's integration with matplotlib provides the flexibility to use it in various environments supported by matplotlib. This includes conducting exploratory data analysis in notebooks, enabling real-time interactions in graphical user interface (GUI) applications, and generating final output in different raster and vector formats.

Matplotlib boasts a comprehensive and powerful API, allowing you to fine-tune every aspect of your figures to suit your preferences. By combining Seaborn's user-friendly high-level interface with matplotlib's deep customizability, you can efficiently explore your data while having the flexibility to create graphics that can be refined into high-quality, publication-ready visuals.


```python
import warnings
import seaborn as sns

# Suppress warnings
warnings.filterwarnings("ignore")

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)

# Apply the default theme
sns.set_theme()

```

***
#### Statistical estimation using Seaborn 
If there is an interest in an average value of a specific variable as a function of other variables, Seaborn has a lot of functions that will do the essential statistical estimation to answer the questions. For example, in the code below:

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
sns.set_theme()
```

***
Another use of the Seaborn library is not just descriptive statistics, however, scatterplots can be improved by including a regression model using a method called **Implot**, for example:

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
sns.set_theme()
sns.set_theme()
```

***
#### Distributional representations
If it is required to know about the distribution of the data set the **displot** library is used to visualise these distributions including histograms and kernel density estimations. For example:

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
sns.set_theme()
```

***
Seaborn functions streamline data visualization by combining multiple plots for efficient insights. For instance, the **jointplot()** explores data relationships, showing joint distributions between variables and their individual marginal distributions. For example:

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
sns.set_theme()
```
